## WEB SCRAPPING Worlcup 2018

In this program we are going to get the general stats of a match of the worldcup 2018 and then do a csv with them.

These are the stats that we are interested in : Posesión, Fueras de juego, Saques de esquina, Duelos ganados, Tiros fuera, Tiros a puerta, Tiros dentro del área, Tiros desde fuera del área, Paradas, Pases, Pases precisos, Faltas, Tarjetas amarillas, Tarjetas rojas.

```python
#we use BeautifulSoup to gte the html of the webpage were we are going to get this data
import requests
from bs4 import BeautifulSoup

#Get the webpage where we get the information
source=requests.get("https://es.besoccer.com/partido/seleccion-rusia/seleccion-arabia-saudi/2018686661")
soup=BeautifulSoup(source.text,"html.parser")
#This make looks better the html
soup=BeautifulSoup(soup.prettify(), "html.parser")
print(soup)
```

```python
team=soup.find_all(class_="name")
team_local=team[0].get_text()[13:-12]
team_visitor=team[1].get_text()[13:-12]
goals_local=int(soup.find("span",class_="r1").get_text())
goals_visitor=int(soup.find("span",class_="r2").get_text())
if goals_local>goals_visitor:
    win_local='Yes'
    win_visitor='No'
if goals_local<goals_visitor:
    win_local='No'
    win_visitor='Yes'
if goals_local==goals_visitor:
   win_local='Tie'
   win_visitor='Tie'

print(goals_local,goals_visitor)
```

```python
#Get the webpage where we get the information
source=requests.get("https://es.besoccer.com/partido/seleccion-rusia/seleccion-arabia-saudi/2018686661/alineaciones")
soup=BeautifulSoup(source.text,"html.parser")
#This make looks better the html
soup=BeautifulSoup(soup.prettify(), "html.parser")
```

```python
formation=soup.find_all("div",class_="tactic")
formation_local=formation[0].get_text()[13:-12]
formation_visitor=formation[1].get_text()[13:-12]
```

## Creating a csv with the data

I create a new csv with the name worldcup2018 and I add all the data that I have collected

```python
import csv
datos = [[team_local,goals_local,win_local,possesion_local,Offside_local,Corners_local,Disputeswon_local,Shots_insideaerea_local,
         Shots_outsideaerea_local,Goalkick_local,Saves_local,Passes_local,Passescorrect_local,Freekicks_local,yellowcards_local,
         redcards_local],
        [team_visitor,goals_visitor,win_visitor,possesion_visitor,Offside_visitor,Corners_visitor,Disputeswon_visitor,Shots_insideaerea_visitor,
        Shots_outsideaerea_visitor,Goalkick_vistor,Saves_visitor,Passes_visitor,Passescorrect_visitor,Freekicks_visitor,
        yellowcards_visitor,redcards_visitor]]

# Open CSV
with open('worldcup2018.csv', mode='w', newline='') as archivo_csv:
    
    # Crea objeto writer
    writer = csv.writer(archivo_csv)
    
    # Escribe encabezados
    writer.writerow(['Team', 'Goals', 'Win','Possesion','Offsides','Corner','Disputes_won','Shots_inside_area','Shots_outside_area','Goal Kick','Saves','Passes','Passes_complete','Faults','Yellow_cards','Red_cards'])
    
    # Escribe datos
    writer.writerows(datos)
    
# Cerrar archivo
archivo_csv.close()
```
