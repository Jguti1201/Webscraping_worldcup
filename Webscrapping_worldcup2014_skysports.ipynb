```python
#we use BeautifulSoup to gte the html of the webpage were we are going to get this data
import requests
from bs4 import BeautifulSoup

#Get the webpage where we get the information
source=requests.get("https://www.skysports.com/football/brazil-vs-croatia/stats/305738")
soup=BeautifulSoup(source.text,"html.parser")
#This make looks better the html
soup=BeautifulSoup(soup.prettify(), "html.parser")
print(soup)
```

```python
goals=soup.find_all(class_="sdc-site-match-header__team-score-block")
goals_local=int(goals[0].get_text())
goals_visitor=int(goals[1].get_text())
if goals_local>goals_visitor:
    win_local='Yes'
    win_visitor='No'
if goals_local<goals_visitor:
   win_local='No'
   win_visitor='Yes'
if goals_local==goals_visitor:
   win_local='Tie'
   win_visitor='Tie'

print(goals_local,goals_visitor)
```

```python
team_local=soup.find(class_="sdc-site-match-stats__team-name").get_text()[13:-12]
team_visitor=soup.find(class_="sdc-site-match-stats__team-name sdc-site-match-stats__team-name--away").get_text()[13:-12]
print(team_local,team_visitor)
```

```python
stats=soup.find_all(class_="sdc-site-match-stats__val")
Possesion_local=float(stats[0].get_text()[15:-14])
Possesion_visitor=float(stats[1].get_text()[15:-14])
Totalshots_local=int(stats[2].get_text()[15:-14])
Totalshots_visitor=int(stats[3].get_text()[15:-14])
Shotsontarget_local=int(stats[4].get_text()[15:-14])
Shotsontarget_visitor=int(stats[5].get_text()[15:-14])
Shotsoftarget_local=int(stats[6].get_text()[15:-14])
Shotsoftarget_visitor=int(stats[7].get_text()[15:-14])
Shotsblocks_local=int(stats[8].get_text()[15:-14])
Shotsblocks_visitor=int(stats[9].get_text()[15:-14])
Passingaccuracy_local=float(stats[10].get_text()[15:-14])
Passingaccuracy_visitor=float(stats[11].get_text()[15:-14])
Clearcutchances_local=int(stats[12].get_text()[15:-14])
Clearcutchances_visitor=int(stats[13].get_text()[15:-14])
Corners_local=int(stats[14].get_text()[15:-14])
Corners_visitor=int(stats[15].get_text()[15:-14])
Offsides_local=int(stats[16].get_text()[15:-14])
Offsides_visitor=int(stats[17].get_text()[15:-14])
Tacklesaccuracy_local=float(stats[18].get_text()[15:-14])
Tacklesaccuracy_visitor=float(stats[19].get_text()[15:-14])
AerialDuelsaccuracy_local=float(stats[20].get_text()[15:-14])
AerialDuelsaccuracy_visitor=float(stats[21].get_text()[15:-14])
Saves_local=int(stats[22].get_text()[15:-14])
Saves_visitor=int(stats[23].get_text()[15:-14])
Foulscommited_local=int(stats[24].get_text()[15:-14])
Foulscommited_visitor=int(stats[25].get_text()[15:-14])
Yellowcards_local=int(stats[26].get_text()[15:-14])
Yellowcards_visitor=int(stats[27].get_text()[15:-14])
Redcards_local=int(stats[28].get_text()[15:-14])
Redcards_visitor=int(stats[29].get_text()[15:-14])
```

```python
import csv
datos=[[team_local,goals_local,win_local,Possesion_local,Totalshots_local,Shotsontarget_local,Shotsoftarget_local,
       Shotsblocks_local,Passingaccuracy_local,Clearcutchances_local,Corners_local,Offsides_local,Tacklesaccuracy_local,
       AerialDuelsaccuracy_local,Saves_local,Foulscommited_local,Yellowcards_local,Redcards_local],
       [team_visitor,goals_visitor,win_visitor,Possesion_visitor,Totalshots_visitor,Shotsontarget_visitor,Shotsoftarget_visitor,
       Shotsblocks_visitor,Passingaccuracy_visitor,Clearcutchances_visitor,Corners_visitor,Offsides_visitor,Tacklesaccuracy_visitor,
       AerialDuelsaccuracy_visitor,Saves_visitor,Foulscommited_visitor,Yellowcards_visitor,Redcards_visitor]]

# Open CSV
with open('worldcup2014_Skysport.csv', mode='w', newline='') as archivo_csv:
    
    # Crea objeto writer
    writer = csv.writer(archivo_csv)
    
    # Escribe encabezados
    writer.writerow(['TEAM','GOALS','WIN','POSSESSION','TOTAL SHOTS','SHOTS ON GOAL','SHOTS OUT','BLOCKED SHOT','PASSING ACCURACY',
                     'CLEAR CUT CHANCES','CORNERS','OFFSIDES','TACKLES ACCURACY','AERIAL DUELS ACCURACY',
                     'SAVES','FOULS COMMITED','YELLOW CARDS','RED CARDS'])
    
    # Escribe datos
    writer.writerows(datos)
    
# Cerrar archivo
archivo_csv.close()
```
